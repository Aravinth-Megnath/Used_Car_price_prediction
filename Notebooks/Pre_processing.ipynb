```python
import pandas as pd
bangalore_df = pd.read_excel(r"C:\Users\Hp\Car\ML ready data\Bangalore_car_detail.xlsx")
chennai_df = pd.read_excel(r"C:\Users\Hp\Car\ML ready data\Chennai_car_detail.xlsx")
delhi_df = pd.read_excel(r"C:\Users\Hp\Car\ML ready data\Delhi_car_detail.xlsx")
hyderabad_df = pd.read_excel(r"C:\Users\Hp\Car\ML ready data\Hyderabad_car_detail.xlsx")
jaipur_df = pd.read_excel(r"C:\Users\Hp\Car\ML ready data\Jaipur_car_detail.xlsx")
kolkata_df = pd.read_excel(r"C:\Users\Hp\Car\ML ready data\Kolkata_car_detail.xlsx")
```

```python
bangalore_df['City'] = 'Bangalore'
chennai_df['City'] = 'Chennai'
delhi_df['City'] = 'Delhi'
hyderabad_df['City'] = 'Hyderabad'
jaipur_df['City'] = 'Jaipur'
kolkata_df['City'] = 'Kolkata'
```

```python
df1 = pd.concat([bangalore_df,chennai_df],ignore_index = True)
df2 = pd.concat([delhi_df,hyderabad_df],ignore_index = True)
df3 = pd.concat([jaipur_df,kolkata_df], ignore_index = True)
df4 = pd.concat([df1,df2],ignore_index = True)
final_df = pd.concat([df4,df3],ignore_index = True)
final_df.shape
```

```python
final_df.City.nunique()
```

```python
final_df.price.dtype
```

```python
lakh_df = final_df[final_df['price'].str.contains('lakh',case=False,na = False)]
crore_df = final_df[final_df['price'].str.contains('Crore',case = False, na=False)]
thousand_df = final_df[final_df['price'].str.contains(',')]
thousand_df['price'],lakh_df['price'],crore_df['price']
```

```python
crore_df.price.isna().sum()
```

```python
def extract_num(x):
    x = x.replace("₹",'').replace('Lakh','').replace(',','').replace('Crore','').strip()
    return x
```

```python
lakh_df.loc[:,'price'] = lakh_df['price'].apply(extract_num)
```

```python
lakh_df.loc[:,'price'] = lakh_df['price'].astype('float')
```

```python
lakh_df.loc[:,'price'] = lakh_df['price']*100000
```

```python
lakh_df.price.max()
```

```python
crore_df.loc[:,'price'] = crore_df['price'].apply(extract_num)
crore_df.loc[:,'price'] = crore_df['price'].astype('float')

```

```python
crore_df.loc[:,'price'] = crore_df['price']*10000000
```

```python
crore_df.price.max()
```

```python
thousand_df.loc[:,'price'] = thousand_df['price'].apply(extract_num)
thousand_df.loc[:,'price'] = thousand_df['price'].astype('float')
```

```python
price_df1 = pd.concat([lakh_df,thousand_df],ignore_index = True)
new_df = pd.concat([price_df1,crore_df])
new_df.shape
```
