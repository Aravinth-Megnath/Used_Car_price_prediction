```python
import pandas as pd
df = pd.read_excel('S:\Car\checking.xlsx')
df.shape
```

```python
df['Comfort & Convenience'].value_counts()
```

```python
df.columns
```

```python
df['Wheel Size'] = df['Wheel Size'].astype(str)
df['Wheel Size'] = df['Wheel Size'].replace(',','',regex=True).replace('m','',regex=True).replace('mm','',regex=True).replace('R','',regex=True)
df['Wheel Size'].value_counts()
```

```python
df['Height'] = df['Height'].astype(str)
df['Width'] = df['Width'].astype(str)
df['Length'] = df['Length'].astype(str)
df['Height'] = df['Height'].replace(',','',regex=True)
df['Width'] = df['Width'].replace(',','',regex=True)
df['Length'] = df['Length'].replace(',','',regex=True)
```

```python
df.drop(columns=['Gear Box'],axis=1,inplace=True)
```

```python
import re
def clean(x):
    pattern = r'^(\d{4})-'
    re.match(pattern, x)
```

```python
df.drop(columns=['Height','Width','Length'],axis=1,inplace=True)
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())
```

```python
df['Gear'] = df['Gear'].fillna(df['Gear'].mode()[0])
```

```python
df.dtypes
```

```python
df.drop(['Wheel Size'],axis=1,inplace=True)
```

```python
#outliers 
from scipy import stats
import numpy as np
# Assuming df is your DataFrame
# Calculate the Z-scores
z_scores = stats.zscore(df)

# Remove outliers (keeping only those with Z-score < 3 in absolute value)
df = df[(np.abs(z_scores) < 3).all(axis=1)]
```

```python
corr_df = df.corr()
corr_df
```

```python
correlation_with_price = corr_df['price_in_lakhs'].sort_values(ascending=False)
correlation_with_price
```

```python
relevant_features = correlation_with_price[correlation_with_price>0.35].index
relevant_features_2 = correlation_with_price[correlation_with_price<-0.35].index
relevant_features_2
```

```python
final_features = relevant_features.append(relevant_features_2)
final_features
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
len(relevant_features)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

df[final_features] = sc.fit_transform(df[final_features])
```

```python
df
```

```python
df.isna().sum()
```

```python
X = df[final_features]
X.drop('price_in_lakhs',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, df['price_in_lakhs'], test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```
